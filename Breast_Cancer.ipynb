```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
data = datasets.load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df.head()
```

```python
# Correlation between each feature
plt.figure(figsize=(14,12))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

<!-- Load the Dataset  -->

```python
df['target'] = data.target
```

```python
df.columns
```

```python
X = df[['mean radius', 'mean texture', 'mean perimeter', 'mean area',
       'mean smoothness', 'mean compactness', 'mean concavity',
       'mean concave points', 'mean symmetry', 'mean fractal dimension',
       'radius error', 'texture error', 'perimeter error', 'area error',
       'smoothness error', 'compactness error', 'concavity error',
       'concave points error', 'symmetry error', 'fractal dimension error',
       'worst radius', 'worst texture', 'worst perimeter', 'worst area',
       'worst smoothness', 'worst compactness', 'worst concavity',
       'worst concave points', 'worst symmetry', 'worst fractal dimension']]
y = df['target']
```

```python
X
```

```python
y
```

```python
# Train Test Split 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=10000)
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
y_pred_lr
```

```python
# Performance Metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, r2_score, mean_absolute_error
print(accuracy_score(y_test, y_pred_lr))
print(confusion_matrix(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))
print(r2_score(y_test, y_pred_lr))
print(mean_absolute_error(y_test, y_pred_lr))
```

```python
# SVC Implementation
from sklearn.svm import SVC
```

```python
# Linear Kernel
lin = SVC(kernel='linear')
lin.fit(X_train, y_train)
y_pred_lin = lin.predict(X_test)
print(accuracy_score(y_test, y_pred_lin))
print(classification_report(y_test, y_pred_lin))
print(confusion_matrix(y_test, y_pred_lin))
print(r2_score(y_test, y_pred_lin))
print(mean_absolute_error(y_test, y_pred_lin))
```

```python
# Polynomial Kernel
poly = SVC(kernel='poly')
poly.fit(X_train, y_train)
y_pred_poly = poly.predict(X_test)
print(accuracy_score(y_test, y_pred_poly))
print(classification_report(y_test, y_pred_poly))
print(confusion_matrix(y_test, y_pred_poly))
print(r2_score(y_test, y_pred_poly))
print(mean_absolute_error(y_test, y_pred_poly))
```

```python
# RBF Kernel
rbf = SVC(kernel='rbf')
rbf.fit(X_train, y_train)
y_pred_rbf = rbf.predict(X_test)
print(accuracy_score(y_test, y_pred_rbf))
print(classification_report(y_test, y_pred_rbf))
print(confusion_matrix(y_test, y_pred_rbf))
print(r2_score(y_test, y_pred_rbf))
print(mean_absolute_error(y_test, y_pred_rbf))
```

```python
# Sigmoid Kernel
sig = SVC(kernel='sigmoid')
sig.fit(X_train, y_train)
y_pred_sig = sig.predict(X_test)
print(accuracy_score(y_test, y_pred_sig))
print(classification_report(y_test, y_pred_sig))
print(confusion_matrix(y_test, y_pred_sig))
print(r2_score(y_test, y_pred_sig))
print(mean_absolute_error(y_test, y_pred_sig))
```

```python
# GridSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C' : [0.1, 1, 10, 100, 1000],
        'gamma' : [1, 0.1, 0.01, 0.001, 0.0001],
        'kernel' : ['rbf','linear','polynomial']}
```

```python
grid = GridSearchCV(SVC(), param_grid=param_grid, refit=True, cv=5, verbose=3)
```

```python
print(grid.fit(X_train, y_train))
```

```python
grid.best_params_
```

```python
y_pred_grid = grid.predict(X_test)
print(accuracy_score(y_test, y_pred_grid))
print(classification_report(y_test, y_pred_grid))
print(confusion_matrix(y_test, y_pred_grid))
```
